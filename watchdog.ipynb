```python
from watchdog.observers import Observer
from watchdog.events import FileSystemEventHandler
import time
```

```python
class MyHandler(FileSystemEventHandler):
    def __init__(self):
        self._last_event = {}

    def on_modified(self, event):
        now = time.time()
        last = self._last_event.get(event.src_path, 0)
        if now - last < 1:  # 1 másodpercen belüli duplikátok kiszűrése
            return
        self._last_event[event.src_path] = now
        print(f"Változás: {event.src_path}")

    def on_created(self, event):
        print(f"Létrehozva: {event.src_path}")

    def on_deleted(self, event):
        print(f"Törölve: {event.src_path}")
```

```python
path = "c:/Users/gallz/Downloads/"  # Itt add meg a figyelt mappa elérési útját
event_handler = MyHandler()
observer = Observer()
observer.schedule(event_handler, path, recursive=True)
observer.start()

try:
    while True:
        time.sleep(1)
except KeyboardInterrupt:
    observer.stop()
observer.join()
```
